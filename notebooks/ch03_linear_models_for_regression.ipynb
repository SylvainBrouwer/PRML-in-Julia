## Linear Regression

```julia
#Create toy data
x = collect(-10:0.5:10)
y = map(p->2p-1, x) + randn(length(x))*2
plot(x,y, seriestype=:scatter, label="y")

#Fit regression
reg = LinearRegression()
fit!(reg, x, y)
y_pred = predict(reg, x)
plot!(x, y_pred, label="regression")
```

## Ridge Regression

```julia
ridge = RidgeRegression(5)
fit!(ridge, x, y)
y_pred = predict(reg, x)
plot(x,y, seriestype=:scatter, label="y")
plot!(x, y_pred, label="ridge")
```
