```julia
include("../src/prml/regression/regression.jl")

using LinearAlgebra
using Plots
using .regression
```

## Linear Regression

```julia
# Create toy data
x = collect(-10:0.5:10)
y = map(p->2p-1, x) + randn(length(x))*2
plot(x,y, seriestype=:scatter, label="y")

# Fit regression
reg = LinearRegression()
fit!(reg, x, y)
y_pred = predict(reg, x)
plot!(x, y_pred, label="regression")
```

## Ridge Regression

```julia
# Create toy data
x = collect(-10:0.5:10)
y = map(p->2p-1, x) + randn(length(x))*2
x_blob = [6, 7.2, 7.5, 7.4, 6.9]
y_blob = [4.2, 4, 4.5, 4.2, 5]
append!(x, x_blob)
append!(y, y_blob)

# Fit regressor
lambda = 100
linear = LinearRegression()
ridge = RidgeRegression(lambda)
fit!(linear, x, y)
fit!(ridge, x, y)

# Predict and plot
y_linear = predict(linear, x)
y_ridge = predict(ridge, x)
plot(x,y, seriestype=:scatter, label="y")
plot!(x, y_linear, label="linear")
plot!(x, y_ridge, label="ridge")
```

## Polynomial Regression

```julia
# Create toy data
x = collect(-10:0.5:10)
y = map(p -> 2 - 2p + p^2 + p^3, x) + randn(length(x))*100

# Fit regressor
polyreg = PolynomialRegression()
fit!(polyreg, x, y)

# Predict and plot
y_pred = predict(polyreg, x)
plot(x, y, seriestype=:scatter, label="y")
plot!(x, y_pred, label="polynomial")
```
